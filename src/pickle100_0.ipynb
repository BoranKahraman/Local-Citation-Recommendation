```python
def window(text, target_str, left_window_size = 150, right_window_size = 100):
    loc = text.find(target_str)
    
    left_threshold = loc - left_window_size
    right_threshold = loc + right_window_size
    
    if (left_threshold < 0):
        left_threshold = 0
    elif (right_threshold > len(text) - 1):
        right_threshold = len(text) - 1
        
    return text[left_threshold: right_threshold]
    
```

```python
def pdf_to_text(pdf_file_path):
    text = ""
    with open(pdf_file_path, 'rb') as pdf_file:
        pdf_reader = PyPDF2.PdfReader(pdf_file)
        for page_num in range(len(pdf_reader.pages)):
            page = pdf_reader.pages[page_num]
            text += page.extract_text()
    text = text.replace('\n', ' ')
    return text
```

```python
pattern_to_remove =   r'\b(\d+)\s+([\w\s]+)\((\d{4})\)\.\s*([^\.]+)\.\s*([^,]+),\s*(\d+)\s*\((\d+)\),\s*(\d+)-(\d+)\b'
pattern_to_remove_reverse = r'\b([\w\s]+?)\s+\((\d{4})\)\.\s*([^\.]+)\.\s*([^,]+),\s*(\d+)\s*\((\d+)\),\s*(\d+)-(\d+)\s+(\d+)\b'





matches_1 = re.finditer(pattern_to_remove, text)
matches_2 = re.finditer(pattern_to_remove_reverse, text)

# Initialize a count variable
count_1 = 0
count_2 = 0


# Iterate through the matches and count them
for match_1 in matches_1:
    count_1 += 1
    matched_sentence = match_1.group()
    print(matched_sentence)
    print("\n--------------\n")

# Print the total count of matching sentences
print(f"Total matches_1 found: {count_1}")

# Iterate through the matches and count them
for match_2 in matches_2:
    count_2 += 1
    matched_sentence_2 = match_2.group()
    print(matched_sentence_2)
    print("\n--------------\n")

# Print the total count of matching sentences
print(f"Total matches_2 found: {count_2}")
```

```python
def split_article(article_text):
    reference_pattern = r"Kaynakça|References|Bibliography|Citations"
    reference_match = re.search(reference_pattern, article_text, re.IGNORECASE)

    if reference_match:
        main_text = article_text[:reference_match.start()] #extract main text from references
        reference_section = article_text[reference_match.start():]

        return main_text, reference_section
    else:
        return article_text, None

text, references = split_article(text)
```

```python
def custom_sentence_split(text):
    sentences = sent_tokenize(text, language="turkish")
    # Post-process the sentences to combine sentences ending with abbreviations
    new_sentences = []
    i = 0

    while i < len(sentences):
        if i < len(sentences) - 1 and sentences[i].endswith('.') and re.match(r'\b[A-Z][a-zA-Z]*\.', sentences[i]):
            new_sentences.append(sentences[i] + ' ' + sentences[i + 1])
            i += 1
        else:
            new_sentences.append(sentences[i])
        i += 1
    return new_sentences

def find_references(text, title, author, year):
    # Define a regular expression pattern for references
    pattern = re.compile(r'\(.*?\d{4}.*?\)')

    # Custom sentence splitting
    sentences = custom_sentence_split(text)

    # Find sentences that contain references
    sentences_with_references = [
        sentence for sentence in sentences
        if pattern.search(sentence) and
           sentence.lower() != title.lower() and
           author not in sentence and
           str(year) not in sentence
    ]
    return sentences_with_references
```
